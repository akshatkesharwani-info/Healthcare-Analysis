 ANOMALY DETECTION

```python
zero_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in zero_cols:
 count = (df[col] == 0).sum()
 print(f"{col}: {count} zero values")
```

```python
# Replace zeros with NaN in medical columns
df[zero_cols] = df[zero_cols].replace(0, np.nan)
```

#5. INFERENTIAL STATISTICS

```python


features = ['Glucose', 'BloodPressure', 'BMI', 'Age', 'Insulin', 'DiabetesPedigreeFunction']
print("\nT-Test Results: Diabetic vs Non-Diabetic")
print("-" * 50)
for feat in features:
 group0 = df[df['Outcome'] == 0][feat]
 group1 = df[df['Outcome'] == 1][feat]
t, p = stats.ttest_ind(group0, group1)
significance = "SIGNIFICANT" if p < 0.05 else "not significant"
print(f"{feat}: p = {p:.4f} → {significance}")
```

INSIGHTS
1. Glucose level is the strongest predictor of diabetes with the highest statistical significance.
2. Diabetic patients have higher average BMI and Age compared to non-diabetic patients.
3. Insulin and Blood Pressure show variation but are less significant predictors.
4. Missing values were identified in medical fields and handled using median imputation.
5. Correlation analysis shows Glucose has the strongest relationship with diabetes outcome.

INTERPRETATION


Higher glucose levels significantly increase the risk of diabetes.
Patients with higher BMI and age are more likely to develop diabetes.
Early monitoring of these indicators can help in prevention and diagnosis.

FINAL CONCLUSION

Glucose, BMI, and Age are the most important indicators for predicting diabetes.
Healthcare professionals should focus on these factors for early detection and treatment planning.
